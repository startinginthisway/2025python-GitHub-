```python
# 字体
import matplotlib.font_manager as font_manager
import numpy as np
font_addres = 'C:\\Windows\\Fonts\\simsun.ttc'
my_font = font_manager.FontProperties(fname=font_addres,size = 10)
```

```python
# 折线图
import matplotlib.pyplot as plt
x = range(20)
y = [np.random.randint(0,30) for i in x]
# y = np.random.randint(0,30,20)
plt.figure(figsize = (20,8),dpi = 80)#设置图片大小和分辨率
plt.title('折线图',fontproperties = my_font)
x_lable = ['']
plt.plot(x,y)

plt.show()
```
